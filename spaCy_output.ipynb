```python
def analyze_job_description(file_path):
  # nlp 생성
  nlp_job_description = spacy.load('/content/drive/MyDrive/spaCy_trained/job_best/content/AI_Resume_Builder/output/model-best')

  # doc 저장
  doc_job_description = fitz.open(file_path)
  text=" "
  for page in doc_job_description:
    text = text+str(page.get_text())
  text = text.strip()
  text = ' '.join(text.split())
  doc_job_description = nlp_job_description(text)

  # return
  dic_job_description = {
      'company': [],
      'project': [],
      'job': [],
      'required skills': [],
      'preferred skills': [],
      'department': [],
      'benefit': [],
      'required year':[],
  }
  for ent in doc_job_description.ents:
    dic_job_description[ent.label_].append(ent.text)
  return dic_job_description
```

```python
def analyze_resume(file_path):
  # nlp 생성
  nlp_resume = spacy.load('/content/drive/MyDrive/spaCy_trained/resume_best/content/AI_Resume_Builder/output_resume/model-best')

  # doc 저장
  doc_resume = fitz.open(file_path)
  text=" "
  for page in doc_resume:
    text = text+str(page.get_text())
  text = text.strip()
  text = ' '.join(text.split())
  doc_resume = nlp_resume(text)

  # return
  dic_resume = {
      'work experience': [],
      'skill': [],
      'college': [],
      'major': [],
      'education': [],
      'ability': [],
      'award': [],
      'age':[],
      'degree':[]
  }
  for ent in doc_resume.ents:
    dic_resume[ent.label_].append(ent.text)
  return dic_resume
```

```python
analyze_jd = analyze_job_description('/content/drive/MyDrive/spaCy_trained/jd test/1.pdf')
analyze_res = analyze_resume('/content/drive/MyDrive/spaCy_trained/res test/1.pdf')
```

```python
def compare_resume_job(dic_resume, dic_job_description):
  list_satisfied = []
  list_desired = []
  
  # skill 요소만 비교
  for ent in dic_job_description['required skills']:
    if ent in dic_resume['skill']:
      list_satisfied.append(ent)
    else:
      list_desired.append(ent)

  for ent in dic_job_description['preferred skills']:
    if ent in dic_resume['skill']:
      list_satisfied.append(ent)
    else:
      list_desired.append(ent)
  
  percentage = len(list_satisfied)/(len(list_satisfied)+len(list_desired))*100
  percentage = round(percentage, 2)

  # 출력
  print("---------------------------------------------------------------------------\n")
  print("Good points, common points in resume and job description are...\n")
  for ent in list_satisfied:
    print(ent)
  print('\n')

  print("---------------------------------------------------------------------------\n")
  print("Desired things, job description requires, but not included in resume are...\n")
  for ent in list_desired:
    print(ent)
  print('\n')

  print("---------------------------------------------------------------------------\n")
  print("The probability that the resume of the user satisfies the job description about skill section is "+str(percentage)+"%!\n")
```

```python
compare_resume_job(analyze_res, analyze_jd)
```
